```python
x = np.linspace(150,210,100)

plot1 = plt.figure(1)
plt.plot(x,stats.norm.pdf(x, 178, 7.7))
plot2 = plt.figure(2)
plt.plot(x,stats.norm.cdf(x, 178, 7.7))
plt.show()
```
