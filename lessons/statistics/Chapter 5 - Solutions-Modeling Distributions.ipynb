### Exercise 5.1 
In the BRFSS (see Section 5.4), the distribution of heights is roughly normal with parameters μ = 178 cm and σ = 7.7 cm for men, and μ = 163 cm and σ = 7.3 cm for women.
In order to join Blue Man Group, you have to be male between 5’10” and 6’1” (see http://bluemancasting.com). What percentage of the U.S. male population is in this range? Hint: use scipy.stats.norm.cdf.

```python
def in_cm(ht):
    return ht*2.54

upper = in_cm(73)
lower = in_cm(70)
print(lower, upper)
```

```python
x = np.linspace(150,210,100)

plot1 = plt.figure(1)
plt.plot(x,stats.norm.pdf(x, 178, 7.7))
plot2 = plt.figure(2)
plt.plot(x,stats.norm.cdf(x, 178, 7.7))
plt.show()
```

```python
#The probability will be the area under the curve between 177.8 and 185.4 cm

answer = stats.norm.cdf(185.42, 178, 7.7) - stats.norm.cdf(177.8, 178, 7.7)
print('Prbability between 5-10 and 6-1 is:', answer)
```
